```python
import os
import glob
```

```python
dirr=os.listdir()
dirr2=glob.glob("*.psd")
```

```python
for ii in dirr2:
    bob2=pd.read_csv(ii, header=None, sep = r"\n",engine="python")
    name=ii.strip(".psd")+'_plain.txt'
    f= open(name,"w+")
    for jj in bob2[0]:
        ptext=re.findall(r' [A-Za-z\',?!.:;-]*\)',str(jj))
        for i in ptext:
            f.write(i.strip("()").strip()+' ')
    f.close()
    
```
